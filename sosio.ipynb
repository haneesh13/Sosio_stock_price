```python
import matplotlib.pyplot as plt
from sklearn import linear_model
import numpy as np
import datetime
```

```python
my = np.asarray(dff)
date_time = []
op = []
hp = []
lp = []
cp = []
for i in range(len(my)):
    date_time.append(my[i][9])
    op.append(my[i][3])
    hp.append(my[i][4])
    lp.append(my[i][5])
    cp.append(my[i][6])

```

```python
def predict_price (datime,prices,ndate):
    linear_mod = linear_model.LinearRegression() 
    datime = np.reshape(datime,(len(datime),1)) 
    prices = np.reshape(prices,(len(prices),1))
    linear_mod.fit(datime,prices) 
    ndate= np.reshape(ndate,(1,-1))
    predicted_price =linear_mod.predict(ndate)
    return predicted_price[0][0]
```

```python
date_entry = input('Enter a date in YYYY-MM-DD-HH-mm-ss format')
year, month, day, hour, minute , seconds  = map(int, date_entry.split('-'))
date_time1= datetime.datetime(year, month, day,hour, minute, seconds) # date at which price has to be predicted

```

```python
date_time1 = time.mktime(date_time1.timetuple())
```

```python
predicted_price = predict_price (date_time,op,date_time1) 
print ("Bar Open price for", date_time1,"is :" ,predicted_price)
predicted_price = predict_price (date_time,hp,date_time1) 
print ("Bar High price for", date_time1,"is :" ,predicted_price)
predicted_price = predict_price (date_time,lp,date_time1) 
print ("Bar Low price for", date_time1,"is :" ,predicted_price)
predicted_price = predict_price (date_time,cp,date_time1) 
print ("Bar Close price for", date_time1,"is :" ,predicted_price)
```
